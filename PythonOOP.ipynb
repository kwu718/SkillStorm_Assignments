```python
class ComputerComponent:

    def __init__(self, manufacturer, model, serial_number):
        self.manufacturer = manufacturer
        self.model = model
        self.serial_number = serial_number
    
    def ComputerComponent(self):
        return(f'Manufacturer: {self.manufacturer}\nModel: {self.model}\nSerial #: {self.serial_number}\n')
    

```

```python
class CPU(ComputerComponent):
    def __init__(self, manufacturer, model, serial_number, cores, clock_speed):
        super().__init__(manufacturer, model, serial_number)
        self.cores = cores
        self.clock_speed = clock_speed
    def over_clock(self, over_clock_speed: int):
        self.clock_speed += over_clock_speed

    def get_type(self):
        return 'CPU'
    
    def __repr__(self):
        return (f'{super().ComputerComponent()}Cores: {self.cores}\nClock Speed: {self.clock_speed}\n')

cpu = CPU("APPLE", "M!", "1231", 2, 5.3)  
```

```python
class Memory(ComputerComponent):
    def __init__(self, manufacturer, model, serial_number, capacity, type):
        super().__init__(manufacturer, model, serial_number)
        self.capacity = capacity
        self.type = type

    def get_type(self):
        return "Memory"
    
    def __repr__(self):
        return (f'{super().ComputerComponent()}Capacity(gb): {self.capacity}\nType: {self.type}\n')
    
mem = Memory('samsung', 'nexus 4', 'adfs', 16, 'DRAM')    
```

```python
class Storage(ComputerComponent):

    curr_sto = 0
    def __init__(self, manufacturer, model, serial_number, storage_type, size):
        super().__init__(manufacturer, model, serial_number)
        self.storage_type = storage_type
        self.size = size

    def get_type(self):
        return 'Storage' 
    
    def almost_full(self):
        return self.curr_sto >= (.9 * self.size)
          
    def __repr__(self):
        return (f'{super().ComputerComponent()}Storage Type: {self.storage_type}\nSize(tb): {self.size}\n')

sto = Storage('Seagate', 'Slim Backup', '654a', 'SSD', 2)
```

```python
class Computer:
    parts_dict = {}
    def __init__(self, obj1, obj2, obj3):
   
        if not all(isinstance(cc, ComputerComponent) for cc in [obj1, obj2, obj3]):
            raise TypeError("Not a Computer Component")
        
        
        self.parts_dict[obj1.get_type()] = obj1
        self.parts_dict[obj2.get_type()] = obj2
        self.parts_dict[obj3.get_type()] = obj3

    
    def display_parts(self):
        for k, v in self.parts_dict.items():
            print(f'{k}\n{v}')
    

    def replace(self, component):
        if type(component) is CPU:
            self.parts_dict['CPU'] = component
        elif type(component) is Memory:
            self.parts_dict['Memory'] = component
        elif type(component) is Storage:
            self.parts_dict['Storage'] = component
        else:
            print("Invalid Component")

    # Does not check for type
    def add_or_replace(self, obj):
        if not isinstance(obj, ComputerComponent):
            print("INVALID Component")
            return
        type_of_obj = obj.get_type()
        self.parts_dict[type_of_obj] = obj
        
       

comp = Computer(cpu, mem, sto)
comp.display_parts()

new_cpu = CPU('Intel', 'i9', 'xc83', 16, '4.2ghz')
new_mem = Memory('MEMCO', 'mem87', 'LL123', 32, 'SRAM')
new_sto = Storage('Sandisk', 'fsdr', '5645la', 'HDD', 10)

comp.add_or_replace(new_cpu)
comp.display_parts()
        
```
